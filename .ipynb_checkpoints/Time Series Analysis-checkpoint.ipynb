```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
series = pd.read_csv('AirPassengers.csv', parse_dates=[0])
```

```python
series.head()
```

```python
print(series.size)
```

```python
series.describe()
```

```python
plt.plot(series['#Passengers'])
plt.show()
```

```python
series['Month'] = pd.to_datetime(series['Month'], format='%Y-%m')
print(series.head())
```

```python
series.index = series['Month']
del series['Month']
print(series.head())
```

```python
plt.figure(figsize=(16,8))
plt.plot(series['Month'], series['#Passengers'])
plt.xlabel('Year')
plt.ylabel('#passengers')
plt.show()
```
